```python
number = 0

# while 반복문 생성
while number <= 10:

    # number 변수 출력
    print(number)

    if number == 3:
        # break로 반복문 강제 종료
        break
    number = number + 1
```

```python
number = 0

# while 반복문 생성
while number <= 10:

    number = number + 1

    if number == 3:
        # continue로 아래 코드 생략
        continue
    print(number)
```

```python
number = 0

# while 반복문 생성
while number <= 10:

    number = number + 1

    if number == 3:
        # pass로 아무런 동작 없이 순차 진행
        pass
    print(number)
```

```python
import random

attempts = 0
secret_number = random.randint(1, 20)

print("1부터 20 사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요: "))
    attempts = attempts + 1

    if guess < secret_number:
        print("더 큰 숫자입니다.")
    elif guess > secret_number:
        print("더 작은 숫자입니다.")
    else:
        print("정답입니다! 시도 횟수: ", attempts, "번")
        break
```

```python
import random

attempts = 5
secret_number = random.randint(1, 10)

print("1부터 10까지의 숫자 중 하나를 맞춰보세요.")
print("총 5번의 기회가 주어집니다.")

for count in range(attempts):
    count = count + 1
    guess = int(input(f"{count}번째 시도: 숫자를 입력하세요: "))

    if guess == secret_number:
        print(f"축하합니다! 숫자 {secret_number}을(를)  맞췄습니다.")
        break
    elif guess < secret_number:
        print("더 큰 숫자입니다.")
    else:
        print("더 작은 숫자입니다.")

    if count == 5:
        print(f"기회를 모두 사용했습니다. 정답은 {secret_number}였습니다.")
```

```python
import random

attempts = 0
secret_number = random.randint(1, 20)

print("1부터 20 사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요: "))
    attempts = attempts + 1

    if guess < 1 or guess > 20:
        print("유효한 숫자가 아닙니다. 1부터 20 사이의 숫자를 입력하세요.")
        continue

    if guess < secret_number:
        print("더 큰 숫자입니다.")
    elif guess > secret_number:
        print("더 작은 숫자입니다.")
    else:
        print("정답입니다! 시도 횟수", attempts, "번")
        break
```
