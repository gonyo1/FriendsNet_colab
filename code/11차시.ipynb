```python
def juice(sugar, *fruits):
    # 음료에 들어갈 설탕의 양
    print("설탕을", sugar, "g 넣습니다.")

    # 음료에 들어갈 과일
    for fruit in fruits:
        print(fruit, "를 넣습니다.")

    # 완성 메세지 출력
    print("음료가 모두 준비되었습니다.")

# 함수 호출
juice(200, "사과", "바나나", "토마토")
```

```python
def juice(sugar, liquid="물", *fruits, **toppings):
    # 음료에 들어갈 설탕의 양
    print("설탕을", sugar, "g 넣습니다.")

    # 음료에 들어가는 액체의 종류
    print(liquid, "을(를) 넣습니다.")

    # 음료에 들어갈 과일
    for fruit in fruits:
        print(fruit, "를 넣습니다.")

    # 음료에 들어갈 토핑
    for topping, value in toppings.items():
        print(topping, "을(를)", value,"g 넣습니다.")

    # 완성 메세지 출력
    print("음료가 모두 준비되었습니다.\n")

# 함수 호출
juice(200, "사과", "바나나", "토마토")
juice(200, "사과", "바나나", "토마토", 초코시럽=30, 초코칩=10)
```

```python
def get_csv_data(*args):
    # 문항 결과를 저장하는 딕셔너리
    result = dict()

    for column in args:
        # 문항별 결과를 저장
        responses = df[column].tolist()

        # 콤마로 구분된 응답결과를 처리하고 저장
        for index, response in enumerate(responses):
            name = names[index]
            # '안종호, 박은지'
            response = response.split(", ")

            # result 딕셔너리에 response 결과를 저장
            result[name] = response

        print(result)

get_csv_data("문항1", "문항2")
```
