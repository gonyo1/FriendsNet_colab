```python
import csv
data = [
       ['이름','성별','번호'],
       ['안종호','남','17'],
       ['김태하','남','5'],
       ['박은미','여',2]
       ]
with open('/content/drive/MyDrive/sample.csv', 'w') as file:
  writer = csv.writer(file)
  writer.writerows(data)
```

```python
data = []
with open('/content/drive/MyDrive/sample.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)
print(data)
header = ["이름", "성별", "번호", "학년"]
data.insert(0,header)
print(data)

with open('/content/drive/MyDrive/sample.csv', 'w') as file:
    writer = csv.writer(file)
    for row in data:
        writer.writerow(row)
```
