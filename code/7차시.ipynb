```python
# 출력: Hello World
print("Hello", "World")  # 기본 sep=' ', end='\n'

# 출력: Hello-World!
print("Hello", "World", sep='-', end='!')

# 파일 output.txt에 "Hello"를 씁니다.
print("Hello", file=open('output.txt', 'w'))
```

```python
names = ["김태하", "안종호", "박은미", "바다"]

for index, name in enumerate(names):
    print(index, name)
```

```python
names = ["김태하", "안종호", "박은미", "바다"]

for index, name in enumerate(names, start=100):
    print(index, name)
```
