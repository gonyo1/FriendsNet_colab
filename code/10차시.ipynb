```python
def change(word):
    upper_word = word.upper()  # 문자열을 대문자로 변환
    lower_word = word.lower()  # 문자열을 소문자로 변환
    return upper_word, lower_word  # 변환된 문자열 변환

word = change("heLLO")
print(word)
```

```python
def calculation(a, b):
    add = a + b
    sub = a - b
    mul = a * b

    if b != 0:
        div = a / b
    else:
        div = "나누는 수가 0이 아닌 값을 입력하세요"

    return add, sub, mul, div

result = calculation(100, 5)
print(result)
```

```python
def 곱셈구구(multiple):
    result = []

    for item in range(1, 10):
        result.append(multiple * item)

    return result

multiple_3 = 곱셈구구(3)
print(multiple_3)
```
