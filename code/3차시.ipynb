```python
print(5 == 10)  # 5와 10은 같다 (거짓)
print(5 != 10)  # 5와 10은 같지 않다 (참)
print(5 < 10)   # 5는 10보다 작다 (참)
print(5 <= 10)  # 5는 10보다 작거나 같다 (참)
print(5 > 10)   # 5는 10보다 크다 (거짓)
print(5 >= 10)  # 5는 10보다 크거나 같다 (거짓)
```

```python
print(not True)     # 출력: False

print(not 10 > 5)   # 출력: False
```
