```python
fruits = ["사과", "바나나", "오렌지"]

# + 연산자 사용
fruits = fruits + ["체리"]
print("(1) + 연산자 사용 결과:", fruits)

# append 함수 사용
fruits.append("배")
print("(2) append 함수 사용 결과", fruits)

# extend 함수 사용
fruits.extend(["포도", "딸기"])
print("(3) extend 함수 사용 결과:", fruits)

# insert 함수 사용
fruits.insert(0, "수박")
print("(4) insert 함수 사용 결과:", fruits)
```

```python
# 리스트에서 "사과"를 삭제합니다.
fruits = ["사과", "바나나", "오렌지", "사과"]
fruits.remove("사과")
print(fruits)

# 두 번째 요소(인덱스 1)를 삭제하고 반환합니다.
fruits = ["사과", "바나나", "오렌지", "사과"]
popped_fruit = fruits.pop(1)
print(popped_fruit)
print(fruits)
```

```python
fruits = ["사과", "바나나", "오렌지", "사과"]
# del 명령어 사용
del fruits[0]
print("del 키워드 사용 결과:", fruits)

# clear 명령어 사용
fruits.clear()
print("clear 함수 사용 결과:", fruits)
```

```python
fruits = {"사과": 100, "바나나": 200, "오렌지": 300}

# "포도" 키-밸류 쌍 추가하기
fruits["포도"] = 400
print(fruits)

# "사과" 키-밸류 쌍 덮어쓰기
fruits["사과"] = 500
print(fruits)
```

```python
fruits = {"사과": 100, "바나나": 200, "오렌지": 300}

# "오렌지" 항목을 삭제하고 반환합니다.
popped_value = fruits.pop("오렌지")
print(popped_value)
print(fruits)

# "바나나" 항목을 삭제합니다.
del fruits["바나나"]
print(fruits)
```
