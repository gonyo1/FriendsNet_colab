```python
class1 = ["태하", "종호"]
class2 = ["은미", "바다"]
student_name = input("학생의 이름을 입력하세요: ")

if student_name in class1:
    print(student_name + " 학생은 6학년 1반입니다.")
elif student_name in class2:
    print(student_name + " 학생은 6학년 2반입니다.")
else:
    print(student_name + " 학생은 6학년 학급에 없습니다.")
```

```python
height = float(input("키(m)를 입력하세요: "))
weight = float(input("몸무게(kg)를 입력하세요: "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("저체중입니다.")
elif bmi < 23:
    print("정상체중입니다.")
elif bmi < 25:
    print("비만 전 단계(과체중 or 위험체중)입니다.")
elif bmi < 30:
    print("1단계비만입니다.")
elif bmi < 35:
    print("2단계비만입니다.")
else:
    print("3단계비만(고도비만)입니다.")
```
