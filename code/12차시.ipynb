```python
import matplotlib.pyplot as plt

# 데이터 생성
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

x = [1, 2, 3, 4, 5]
y_line = [1, 4, 2, 5, 3]
y_scatter = [5, 3, 4, 2, 1]

# 2x2 서브플롯 생성
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 첫 번째 서브플롯 - 막대그래프
axs[0, 0].bar(categories, values, color='r')
axs[0, 0].set_title('Bar Graph')

# 두 번째 서브플롯 - 선그래프
axs[0, 1].plot(x, y_line, 'g')
axs[0, 1].set_title('Line Graph')

# 세 번째 서브플롯 - 산점도그래프
axs[1, 0].scatter(x, y_scatter, color='b')
axs[1, 0].set_title('Scatter Plot')

# 네 번째 서브플롯 - 원그래프
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', colors=['r', 'g', 'b', 'c'])
axs[1, 1].set_title('Pie Chart')

# 레이아웃 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1행 2열로 배열된 서브플롯을 생성하고, 피겨 사이즈를 설정합니다.
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# 데이터 설정
categories = ["종호","태하","은미"]
values = [3,2,3]

x = [1, 2, 3, 4, 5]
y_line = [4,6,1,3,6]
y_scatter = [2, 3, 5, 7, 11]

# 첫 번째 서브플롯에 막대 그래프 추가
axs[0,0].bar(x, y, color = ['r','g','b','y'])
axs[0,0].set_title('막대 그래프')
axs[0,0].set_xlabel('X 축')
axs[0,0].set_ylabel('Y 축')

# 두 번째 서브플롯에 산점도 추가
axs[0,1].scatter(x, y_scatter, color='r')
axs[0,1].set_title('산점도')

# 세 번째 서브플롯에 선그래프
axs[1,0].plot(x,y_line, linestyle='--')
axs[1,0].set_title('선그래프')

# 네 번째 서브플롯에 원그래프
axs[1,1].pie(values, labels=categories, autopct='%1.1f%%', colors = ['r','g','b'])
axs[1,1].set_title('원그래프')


# 그래프를 화면에 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

# 나눔 폰트 다운로드 및 설정
os.system("apt-get install -y fonts-nanum")
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
new_font = fm.FontEntry(fname=font_path, name="나눔고딕")
fm.fontManager.ttflist.append(new_font)
plt.rcParams.update({'font.size': 10, 'font.family': '나눔고딕'})

#데이터입력
months = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
avg_temp = [-1.5, -1.9, 4.2, 8.8, 14.6, 19.7, 23.9, 24.3, 20.2, 11.5, 2.9, -2.5]
#Figure객체 생성
fig = plt.figure(figsize=(10,5))
plt.plot(months, avg_temp,linestyle='-')
plt.title("월별 평균 기온 변화")
plt.xlabel("월")
plt.ylabel("평균 기온 (°C)")
plt.show()
```
