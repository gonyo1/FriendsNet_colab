```python
import pandas as pd

# 리스트를 사용하여 시리즈 생성
data_list = [10, 20, 30, 40, 50]
series_from_list = pd.Series(data_list)
print(series_from_list)

```

```python
# 딕셔너리를 사용하여 시리즈 생성
data_dict = {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50}
series_from_dict = pd.Series(data_dict)
print(series_from_dict)
```

```python
# 이중 리스트를 사용하여 데이터프레임 생성
data_list_of_lists = [
    [12, '안종호', 35],
    [21, '김태하', 31],
    [13, '박은미', 30]
]
df_from_list_of_lists = pd.DataFrame(data_list_of_lists, columns=['번호', '이름', '나이'])
print(df_from_list_of_lists)
```
