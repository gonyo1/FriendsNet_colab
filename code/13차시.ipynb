```python
# Pandas 로 데이터 불러오기
import pandas as pd
csv_path = "파이썬으로 레벨 업/data.csv"
csv_df = pd.read_csv(csv_path)
```

#### **14-2-1. CSV 조작하기**

##### 2-1) 데이터 프레임 자료 조작

```python
# CSV 열 이름 확인하기
columns = csv_df.columns.tolist()
print(columns)
```

```python
# CSV 파일에서 자료 변형하기
columns = csv_df.columns.tolist()
name_index = columns.index("2. 이름을 입력해주세요.")
print(name_index)
```

```python
# DataFrame에서 문항별 응답 결과를 추출하기
# 이름보다 한 인덱스 뒤에 있는 열 데이터 가져오기
values = csv_df[columns[name_index + 1]].tolist()
print(values)
```

```python
def get_column_data(add):
    values = csv_df[columns[name_index + add]].tolist()
    return values

names = get_column_data(add=0)
print(names)
```

```python
# csv_data 딕셔너리에 이름/성별 추가하기
csv_data = {}

csv_data["names"] = get_column_data(add=0)
csv_data["gender"] = get_column_data(add=1)

print(csv_data["names"])
print(csv_data["gender"])
```

##### 2-1) 응답 결과 처리

```python
# 하나의 문항 응답 결과 데이터를 저장
result = {}

# 예: 문항 1에 대한 결과를 얻기
values = get_column_data(1 + 1)

# 응답 결과를 반복문으로 처리
for idx_answer, answer in enumerate(values):
    # 응답학생 추출하기
    name = csv_data["names"][idx_answer]

    # 응답결과를 리스트로 변환
    names = answer.split(", ")

    # result 딕셔너리에 name을 Key로, names 를 Value로 하는 쌍 추가하기
    try:
        result[name] += names
    except:
        result[name] = names

    # 중복값 처리
    result[name] = list(set(result[name]))

print(result)
```

```python
def get_values(question, result):
    # 예: 문항 1에 대한 결과를 얻기
    values = get_column_data(question + 1)

    # 응답 결과를 반복문으로 처리
    for idx_answer, answer in enumerate(values):
        # 응답학생 추출하기
        name = csv_data["names"][idx_answer]

        # 응답결과를 리스트로 변환
        names = answer.split(", ")

        # result 딕셔너리에 name을 Key로, names 를 Value로 하는 쌍 추가하기
        try:
            result[name] += names
        except:
            result[name] = names

        # 중복값 처리
        result[name] = list(set(result[name]))

get_values(2, result)
print(result)
```

##### 2-3) 두 개 이상의 응답 결과 처리

```python
# 유사 문항을 그룹으로 처리
groups = [1, 2]

# result 에유사 문항 결과를 저장
result = {}

for idx, question in enumerate(groups):
    get_values(question, result)

print(result)
```

```python
def get_category_values(groups):
    # result 딕셔너리에 두 문항 응답 결과를 저장
    result = {}

    for question in groups:
        get_values(question, result)

    return result

get_category_values([4, 5])
```

```python
# CSV data에 값 저장하기
csv_data["인기스타"] = get_category_values([1, 2])
csv_data["위기학생"] = get_category_values([3])
csv_data["똑똑이"] = get_category_values([4, 5])
csv_data["미래확장"] = get_category_values([6, 7, 8, 9])

print(csv_data)
```

```python
def get_csv_data(**kwargs):
    # CSV 파일 불러오기
    csv_df = pd.read_csv(csv_path)
    columns = csv_df.columns.tolist()
    name_index = columns.index("2. 이름을 입력해주세요.")

    # CSV data에 따라 이름/성별 리스트 추가하기
    csv_data = {}

    # CSV data에 키워드로 받은 유사 문항을 정리하여 추가
    for key, value in kwargs.items():
        csv_data[key] = get_category_values(value)
    csv_data["names"] = get_column_data(add=0)
    csv_data["gender"] = get_column_data(add=1)

    return csv_data
```
